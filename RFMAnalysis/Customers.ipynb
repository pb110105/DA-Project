RFM Analysis

Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load Data

```python
customers = pd.read_csv("data/olist_customers_dataset.csv")
```

Sample Customers data

```python
customers.head(5)
```

Counting The Number Of Customer By State

```python
number_of_customer_by_state = customers.groupby("customer_state").size()
print(number_of_customer_by_state)
ax = number_of_customer_by_state.plot(kind= 'bar', figsize =(12,6), color='skyblue')
for container in ax.containers:
    ax.bar_label(container, padding = 3, fontsize = 10)
plt.title("The Number Of Customer By State", fontsize = 18)
plt.xlabel("States", fontsize = 15)
plt.ylabel("The number of Customers", fontsize = 15)
plt.tight_layout()
plt.show()
```

Top 10 Cities That Customers live in the SP

Top 10 Cities Where Customers Live

```python
top_10_cities = customers['customer_city'].value_counts().head(10).sort_values()
ax = top_10_cities.plot(kind = 'barh', figsize = (12,6), color = 'lightseagreen')
for container in ax.containers:
    ax.bar_label(container, padding = 3, fontsize = 10)
plt.title("Top 10 Cities With The Largest Customer Base")
plt.xlabel("The Number Of Customers")
plt.ylabel("Cities")
print(top_10_cities)
```
